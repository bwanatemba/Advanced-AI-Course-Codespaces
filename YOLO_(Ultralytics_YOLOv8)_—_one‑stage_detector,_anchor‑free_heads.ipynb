```python
# Colab setup (run once)
# !pip -q install ultralytics

from ultralytics import YOLO
import matplotlib.pyplot as plt
import numpy as np
from PIL import Image
import urllib.request, os

# 1) Get a sample image
url = "https://ultralytics.com/images/bus.jpg"
img_path = "bus.jpg"
if not os.path.exists(img_path):
    # Add a User-Agent header to mimic a browser request
    opener = urllib.request.build_opener()
    opener.addheaders = [('User-agent', 'Mozilla/5.0')]
    urllib.request.install_opener(opener)
    urllib.request.urlretrieve(url, img_path)


# 2) Load a small, fast pretrained model (COCO)
model = YOLO("yolov8n.pt")  # n/t/s/m/l/x variants; n = nano, fastest

# 3) Inference
results = model(img_path)   # runs on CPU by default

# 4) Visualize detections
annotated = results[0].plot()  # numpy array with boxes/labels drawn
plt.figure(figsize=(10,6))
plt.imshow(annotated)
plt.axis("off")
plt.title("YOLOv8 detections")
plt.show()

# Optional: access raw boxes/scores/labels
# boxes = results[0].boxes.xyxy.cpu().numpy()
# confs = results[0].boxes.conf.cpu().numpy()
# clss  = results[0].boxes.cls.cpu().numpy()
```
