```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(0)
X = np.random.rand(100, 1)
y = 3 * X.squeeze() + 2 + np.random.randn(100) * 0.1

Xb = np.c_[np.ones((100, 1)), X]
theta = np.linalg.inv(Xb.T @ Xb) @ Xb.T @ y
print("θ (bias, weight):", theta.round(3))

plt.scatter(X, y, s=12, label="Data")
plt.plot([0,1], [theta[0] + theta[1]*0, theta[0] + theta[1]*1], color="red", label="Fit")
plt.title("Linear Regression (Closed-form)")
plt.legend()
plt.show()
```
